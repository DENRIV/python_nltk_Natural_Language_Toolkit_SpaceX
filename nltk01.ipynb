```python
# urllib module will help us to crawl the webpage 
import urllib.request
response =  urllib.request.urlopen('https://en.wikipedia.org/wiki/SpaceX')
html = response.read()
print(html)
```

```python
#  We will use beautiful soup to clean our webpage text of HTML tags.
from bs4 import BeautifulSoup
soup = BeautifulSoup(html,'html5lib')
text = soup.get_text(strip = True)
print(text)
```

```python
# convert the text into tokens.
tokens = [t for t in text.split()]
print(tokens)
```

```python
# Count word Frequency

from nltk.corpus import stopwords
sr= stopwords.words('english')
clean_tokens = tokens[:]
for token in tokens:
    if token in stopwords.words('english'):        
        clean_tokens.remove(token)
freq = nltk.FreqDist(clean_tokens)
for key,val in freq.items():
    print(str(key) + ':' + str(val))
# freq.plot(20, cumulative=False)
```

```python
freq.plot(20, cumulative=False)
```

```python
# frequency items to list

list = []
for key,val in freq.items():
    # print(str(key) + ':' + str(val))
    list.append( [ str(key), val ] ) 
print(list)    
```

```python
# Theory
# Sort a list according to the second element in sublist
def Sort1(sub_li): 
    l = len(sub_li) 
    for i in range(0, l): 
        for j in range(0, l-i-1): 
            if (sub_li[j][1] > sub_li[j + 1][1]): 
                tempo = sub_li[j] 
                sub_li[j]= sub_li[j + 1] 
                sub_li[j + 1]= tempo 
    return sub_li 
  
# Driver Code 
sub_li =[['rimimbi', 10], ['akash', 5], ['ramdom', 20], ['gulliver', 15]] 
print(Sort1(sub_li))
```

```python
# Sort List
print(Sort1(list))
```

```python
# Top List

# sequence[start:stop:step]
totLastItems = 3
toplist = list[-1*totLastItems:]

print(toplist)
```

```python
# Theory
import nltk
from nltk.corpus import brown

# classify a word
world = 'launch'
result = nltk.FreqDist(t for w, t in brown.tagged_words() if w.lower() == world)
print(result.most_common())
```

```python
# Looking for the main Word
import nltk
from nltk.corpus import brown

mainWord = ""

for x in toplist:
    print(x[0])
    
    word = x[0]
    result = nltk.FreqDist(t for w, t in brown.tagged_words() if w.lower() == word)

    if word[0].isupper():
        print("->Capital")
        mainWord = word
        
    res = result.most_common()    
    if len(res) > 0:
        # print(res[0])
        typeWord = res[0][0]
        print(typeWord)
        if typeWord == "NN":
            mainWord = word
    print(res)

print("mainWord: ",mainWord)    
```

```python
# Other way : Looking for the main Word

mainWord = ""

s = ""
for x in toplist:
    s = s + " " + x[0]
print(s)

text = nltk.word_tokenize(s)
resultList = nltk.pos_tag(text)
print(resultList)

for t in resultList:
    word     = t[0]
    typeWord = t[1]
    print(typeWord)
    if typeWord == "NN" or typeWord == "NNP" :
        mainWord = word
    
print("mainWord: ",mainWord)    
```
